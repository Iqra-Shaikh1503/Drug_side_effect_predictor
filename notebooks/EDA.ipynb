VISUALIZATION OF THE DATA

# 1. Correlation matrix

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['rating', 'no_of_reviews']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

# 2. Top 10 most reviewed drug

```python
# top drugs values 
top_drugs = df.groupby("drug_name")['no_of_reviews'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_drugs.values, y=top_drugs.index)
plt.title("Top 10 Most Reviewed Drugs")
plt.xlabel("Number of Reviews")
plt.ylabel("Drug Name")
plt.show()
```

**Conclusion:**
Phentermine has the most reviews, followed by bupropion/naltrexone and Contrave  indicating these are the most commonly used or discussed drugs in the dataset.


# 3. Distribution of ratings

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['rating'], bins=20, kde=True)
plt.title("Distribution of Drug Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

**Conclusion:**
Most drugs received **high ratings**, especially **10**, indicating overall **positive user experiences** with the medications.


# 4. Top 10 medical_conditions

```python
top_conditions = df['medical_condition'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_conditions.values, y=top_conditions.index)
plt.title("Top 10 Treated Conditions")
plt.xlabel("Number of Entries")
plt.ylabel("Medical Condition")
plt.show()
```

**Conclusion:**
Pain, colds & flu, and acne are the most commonly treated conditions, highlighting these as key focus areas in the dataset.


# 4. Side effect word cloud

```python
from wordcloud import wordcloud
text = " ".join(df['side_effects'].dropna().tolist())

wc = WordCloud(width=1000, height=500, background_color='white').generate(text)

plt.figure(figsize=(15, 7))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.title("Common Side Effects (Word Cloud)")
plt.show()
```
